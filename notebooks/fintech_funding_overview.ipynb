```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../fintech_funding_trends.csv")
df.head()
```

```python
pivot = df.pivot_table(
    index="Quarter",
    columns="Region",
    values="Funding_MUSD",
    aggfunc="sum"
)

pivot.plot(marker="o", figsize=(10, 5))
plt.title("Quarterly Fintech Funding by Region")
plt.ylabel("Funding (Million USD)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
